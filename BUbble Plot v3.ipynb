```python

# libraries
import os
from matplotlib import font_manager as fm, rcParams
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.cm as cm
%matplotlib inline

matplotlib.rc('xtick', labelsize=10) 
matplotlib.rc('ytick', labelsize=10) 

import matplotlib.pyplot as plt
 
# specify the custom font to use
#plt.rcParams['font.family'] = 'sans-serif'
#plt.rcParams['font.sans-serif'] = 'Arial'

#fonts
fpath = os.path.join(rcParams["datapath"], "C:\\Users\\danica\\Documents\\holonIQ\\fonts\\Gilroy-Light.otf")
lpath = os.path.join(rcParams["datapath"], "C:\\Users\\danica\\Documents\\holonIQ\\fonts\\Gilroy-ExtraBold.otf")

prop = fm.FontProperties(fname=fpath)
footnote = fm.FontProperties(fname=fpath)
footnote.set_size(7)
large = fm.FontProperties(fname=lpath)
large.set_size(13)
large.set_weight('bold')
```

```python
#read dataset
data = pd.read_csv('C:\\Users\\danica\\Documents\\holonIQ\\datasets\\danica_clean2.csv')


data.dropna(inplace=True)

N=108
country = ['Australia','Brazil','Canada','China','Egypt','France',
          'Hong Kong','India','Japan','Malaysia','New Zealand',
          'Saudi Arabia','Singapore','South Africa','Spain',
           'Sweden','United Kingdom','United States'] 
size = data.groupby('Country').size()
ytd = data.groupby('Country')['YTD Performance'].sum()
net = data.groupby('Country')['Net Margin (TTM)'].sum()

# Choose some random colors
colors=cm.rainbow(np.random.rand(N))

d2 = pd.DataFrame(dict(x=net, y=ytd, s=size, users=country))

fig, ax = plt.subplots(facecolor='w')


ax.scatter(net,ytd, s=size*200, alpha=0.5, c=colors)

for i, txt in enumerate(country):
    ax.annotate(txt, (net[i],ytd[i]))

plt.title('Share Price vs Net Margin')
ax.set_xlabel('Net Margin')
ax.set_ylabel('YTD')
plt.savefig('The cornerstone.png', bbox_inches='tight')
plt.show()
```
